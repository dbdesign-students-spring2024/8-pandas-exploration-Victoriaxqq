In order to indicate clearly, we combine the unit and the value by multiplying 0.01 to every value.

```python
data['Value'] /= 100
data.head(5)
```

According to data inspection, we delete all repeating and meaningless columns including "IND", "SEX", "AGE", "TIME", "PowerCode Code", "PowerCode", "Reference Period Code", "Reference Period", "Flags". Since we change the value into the percentage format, we can delete the unit column called "Unit", "Unit Code".

```python
columns_to_drop = ["IND", "SEX", "AGE", "TIME", "Unit", "Unit Code", 
                   "PowerCode Code", "PowerCode", "Reference Period Code", 
                   "Reference Period", "Flags"]

data.drop(columns=columns_to_drop, inplace=True)
data.head(5)
```

Nowadays, we think that unempolyment rate is more significant than employment rate, so we decide to delete all data about employment rate and the column called "Indicator" to distinguish unempolyment rate or employment rate.

```python
data = data[~data['Indicator'].str.contains('Employment')]
data.drop(columns="Indicator", inplace=True)
data.head()
```

We find that there is some overlapping in the column called "Age Group". The category "15-64" contains the category "15-24", "25-54", and "55-64". Therefore, we need to delete the category "15-64".

```python
data = data[data['Age Group'] != '15-64']
data.head()
```

### Analysis 1 - Overall Unemployment Rate by Age Group In the US

Firstly, we decided to calculate average unemployment rate for each age group in the United States. We chose the United States because it has relatively completed data comparing to other contries. We use `groupby` and `mean` to achieve our goal. 

**Analysis**: The 15-24 age group has the highest unemployment rate. We believe this is due to a lack of educational background among people in this age group, making it harder for them to find jobs. The 25-54 age group has a relatively similar unemployment rate compared to the 55-64 age group.

```python
country_distribution = data['Country'].value_counts()

# Filter the data for the United States
us_data = data[data['Country'] == 'United States']

# Calculate the average unemployment rate for each age group in the United States
us_age_average_unemployment_rate = us_data.groupby('Age Group')['Value'].mean().reset_index()

us_age_average_unemployment_rate
```

### Analysis 2 - Overall Unemployment Rate by Sex In Canada

According to data inspection, the dataset is comparably completed about Canada, so we choose Canada to analyze the unemployment rate about different genders with all years and all ages. Firstly, we extract all rows about Canada and name them "data2". Then, we use `groupby` function to calculate the mean of different sexes.

```python
data2 = data[data['Country'] == 'Canada']
Canada_Sex = data2.groupby("Sex")['Value'].mean()
Canada_Sex
```

### Analysis 3 - Change in Unemployment Rate Over Time for Korea

According to data inspection, the dataset is comparably completed about Korea, so we choose Korea to explore the change of unemployment rate in five different years, 2019, 2020, 2021, 2022, and 2023. After we use `groupby` to calculate the mean of different time in Korea, we find there are lots of data lost between 2010 and 2019, so we exclude 2010 in consideration. 

**Analysis**:We figure out the unemployment rate change during the time.

```python
data3 = data[data['Country'] == 'Korea']
Korea_Time = data3.groupby("Time")['Value'].mean()
Korea_Time
```

```python
dictionary = {
    'Quarter': ['Q1-2019', 'Q1-2020', 'Q1-2021', 'Q1-2022', 'Q1-2023',
                'Q2-2019', 'Q2-2020', 'Q2-2021', 'Q2-2022', 'Q2-2023',
                'Q3-2019', 'Q3-2020', 'Q3-2021', 'Q3-2022', 'Q3-2023',
                'Q4-2019', 'Q4-2020', 'Q4-2021', 'Q4-2022'],
    'Value': [0.057000, 0.052500, 0.060500, 0.042500, 0.037333,
              0.060667, 0.061333, 0.054833, 0.042167, 0.033667,
              0.054000, 0.059000, 0.043500, 0.038833, 0.034167,
              0.053333, 0.060333, 0.044833, 0.036833]
}

df = pd.DataFrame(dictionary)

df['Year'] = df['Quarter'].str.split('-').str[1]
df['Quarter'] = df['Quarter'].str.split('-').str[0]
df.head()
```

### Analysis  4 - Overall Unemployment Rate by Country

We will calculate the overall unemployment rate for each country by averaging the unemployment rates across all age groups and time periods. We use `groupby`, `mean`, and `sort` function to achieve our goal.

**Analysis**: We found that Spain, Greece have relatively high unemployment around 20%.

```python
# Group by country and calculate the mean unemployment rate
overall_unemployment_rate = data.groupby('Country')['Value'].mean().sort_values(ascending=False)

overall_unemployment_rate
```

### Analysis 5: Comparison of Unemployment Rates Between Two Countries

We will compare the unemployment rates between two countries, Spain and Italy, over time. From the data, we observe that Spain consistently has a higher unemployment rate than Italy across all quarters.

```python
# Filter the data for Spain and Italy
spain_italy_data = data[(data['Country'].isin(['Spain', 'Italy'])) ]

spain_italy_data1 = spain_italy_data.groupby(['Country', 'Time'])['Value'].mean().unstack()

print(spain_italy_data1)
```

### Analysis 3 - Change in Unemployment Rate Over Time for Korea

For the third analysis, we draw a time series plot to show the change over time. And by drawing five years in one figure and coloring them uniquely, we can compare them easily and see the tendency. In general, the unemployment is decreasing after 2020 since the damage on economy due to covid is gradually recovered. And in the five years, the unemployment reached climax in 2020 due to covid.

```python
plt.figure(figsize=(10, 6))
for year in df['Year'].unique():
    subset = df[df['Year'] == year]
    plt.plot(subset['Quarter'], subset['Value'], label=year)

plt.title('Quarterly Values Over Time')
plt.xlabel('Quarter')
plt.ylabel('Value')
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Analysis 4 - Box Plot of Unemployment Rates by Country

The box plot above visually represents the distribution of unemployment rates for different countries. Each box represents the interquartile range (IQR) of the unemployment rates, with the median indicated by a line inside the box. Whiskers extend from the box to the minimum and maximum values, excluding any potential outliers.

**Analysis**: Most countries have a similar range of unemployment rates, with some countries (like Spain, and Greece) having higher rates. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot a box plot of unemployment rates by country
plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='Value', data=data)
plt.title('Box Plot of Unemployment Rates by Country')
plt.xlabel('Country')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```
